```python
import os
import pandas as pd
import json

ocr_folder_path = r"C:\d drive\Director Project\Chargrid VQA\ocr_results"
target_folder_path = "C:\d drive\Director Project\Chargrid VQA\ocr_csv_folder"
for json_file in os.listdir(ocr_folder_path):
    final_path = os.path.join(ocr_folder_path, json_file)
    f = open(final_path)
    data = json.load(f)
    bounding_box = []
    text = []
    for i in data['recognitionResults'][0]['lines']:
        for j in i['words']:
            bounding_box.append(j["boundingBox"])
            text.append(j["text"])
    df = pd.DataFrame(list(zip(bounding_box, text)), columns=["Bounding Boxes", "Text"])
    # json_file.replace("json", "csv")
    target_name = json_file.split(".")[0] + ".csv"
    df.to_csv(os.path.join(target_folder_path, target_name), index=False)
```
